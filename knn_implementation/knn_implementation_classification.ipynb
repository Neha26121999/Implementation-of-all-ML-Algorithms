# Knn Classification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('data_cleaned.csv')
```

```python
df.head()
```

```python
x=df.drop('Survived',axis=1)
y=df['Survived']
```

```python
x.shape,y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
col=MinMaxScaler()
x_scaled=col.fit_transform(x)
```

```python
x=pd.DataFrame(x_scaled,columns=x.columns)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=56)
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import f1_score
col=KNN(n_neighbors=8)
col.fit(x_train,y_train)
```

```python
y_pred=col.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

# Elbow For Classifier

```python
def Elbow(K):
    #initiating empty list
    test_error = []
   
    #training model for evey value of K
    for i in K:
        #Instance oh KNN
        clf = KNN(n_neighbors = i)
        clf.fit(x_train, y_train)
        # Appending F1 scores to empty list claculated using the predictions
        tmp = clf.predict(x_test)
        tmp = f1_score(tmp,y_test)
        error = 1-tmp
        test_error.append(error)
    
    return test_error
```

```python
k = range(6, 20, 2)
```

```python
test=Elbow(k)
```

```python
# plotting the Curves
plt.plot(k, test)
plt.xlabel('K Neighbors')
plt.ylabel('Test error')
plt.title('Elbow Curve for test')
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import f1_score
col=KNN(n_neighbors=12)
col.fit(x_train,y_train)
```

```python
# Predicting over the Train Set and calculating F1
test_predict = col.predict(x_test)
k = f1_score(test_predict, y_test)
print('Test F1 Score    ', k )
accuracy_score(y_test,test_predict)
```
