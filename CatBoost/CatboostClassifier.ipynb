```python
import pandas as pd
import numpy as np
df=pd.read_csv("train.csv")
```

```python
df.dropna(inplace=True)
df.drop("Loan_ID",axis=1,inplace=True)

x=df.drop("Loan_Status",axis=1)
y=df['Loan_Status']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=46)
```

```python
from catboost import CatBoostClassifier
# for regression CatBoostRegression
col=CatBoostClassifier()
#categorical_feature=np.where(df.dtypes!=np.float)[0]
col.fit(x_train,y_train,cat_features=([0,1,2,3,4,10]),eval_set=(x_test,y_test))
```

```python
col.score(x_test,y_test)
```

```python
col.score(x_train,y_train)
```

```python
y_pred=col.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
